```python
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
y_train = np.load('train/labelIndex.npy')
y_val = np.load('val/labelIndex.npy')
y_test = np.load('test/labelIndex.npy')
```

```python
np.sum([np.unique(y, return_counts=True)[1] for y in [y_train, y_val, y_test]], axis=0)
```

```python
base_dir = '/data2/nathan/embryo/data/'
plist = []
for f in os.listdir(base_dir):
    if f[:29] not in plist:
        plist.append(f[:29])
y_index = np.load(base_dir + plist[0] + 'labelRaw.npy')
for i in range(1, len(plist)):
    if os.path.isfile(base_dir + plist[i] + 'labelRaw.npy'):
        y_imm = np.load(base_dir + plist[i] + 'labelRaw.npy')
        y_index = np.vstack([y_index, y_imm])
```

```python
len(y_index)
```

```python
sums = np.sum(y_index, axis=0)
```

```python
matplotlib.rcParams.update({'font.size': 40})

```

```python
fig, ax = plt.subplots()
sns.set_context("poster")
sns.set_style("white")
labels = ['tStart', 'tPNF', 't2', 't3', 't4', 't5', 't6', 't7', 't8', 't9+', 'tM', 'tSB', 'tB', 'tEB', 'tHB']
ax.barh(np.arange(len(labels)), sums[::-1], align='center')
ax.set_xlabel("# of frames", fontsize=20)
ax.set_yticks(np.arange(len(labels)))
ax.set_yticklabels(labels[::-1])
ax.set_ylim(-0.75, 14.75)
ax.tick_params(labelsize=20)
plt.show()
```
